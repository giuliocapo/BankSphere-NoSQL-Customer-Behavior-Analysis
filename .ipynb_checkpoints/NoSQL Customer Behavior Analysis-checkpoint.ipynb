# NoSQL Customer Behavior Analysis

## Why MongoDB?

### Flexible Structure and Nested Fields

A document-based database sush as MongoDB is perfect fo nested data because it natively support embedded documents. For example in my project after analyzed the dataset I found that `Income_category` field was filled with some strings referred to the income boundaries of the client. Thanks to MongoDB for a better analysis I could define a nested subdocument called `Income_Info` with fields `Category_Code`, `Min`, and `Max` and so, instead of maintaining multiple rigid tables or performing complex joins, I could keep logically related fields togheter in a single document.

### Schema Validation with Adaptable Schema

MongoDB lets you enforce data quality through schema validation while still keeping its inherent flexibility. In this project, i've set up validation schema for both **Customer** and **Transaction** to ensure that fields were alligned to specific data typer and some really important fields instead were be required. This because is really useful as approach because give us the possibility to have different document with different structures in the same collection and also it's really easy to update in the case we need to add more datas and we want a different approch.

### Unified Data Organization and Analysis

In this project, I split the dataset into two separate collections **customers** transactions** using `CLIENTNUM` as the reference key. This model helps me to mantain distinct entities, making easy to create focused queries.Additionally, MongoDB powerful aggregation pipeline gave me the possibility to perform complex operations on datas offering great flexibility and also making data integration more intiutive (e.g `$lookup`) compared to relational database where rigid schemas and complex joins slow down the analisys.

### Ease of Integration with Python (pymongo) and scalable

In your Jupyter notebook, using python, you directly used **pymongo** to insert, query, and aggregate data in MongoDB. Additionally, even if this is not the case, MongoDB give us the possibility to let the model scale through horizontal scaling and sharding, enabling it to handle high volumes of reads and writes ensuring that your data model can grow as needed.

## 2. Mapping Definition

## 3. Application of Mapping to the DataFrame.
Let's implement the mapping using a custom function and create new columns to represent the numerical values.

```python
# Definition of the updated mapping function
def map_income_category(income):
    """
    Maps the 'Income_Category' to numerical codes and defines income ranges.
    
    Parameters:
    income (str): The income category string.
    
    Returns:
    dict: A dictionary with 'Income_Min', 'Income_Max', and 'Income_Category_Code'.
    """
    if income == 'less than $40k':
        return {'Income_Min': 0, 'Income_Max': 40000, 'Income_Category_Code': 1}
    elif income == '$40k - $60k':
        return {'Income_Min': 40000, 'Income_Max': 60000, 'Income_Category_Code': 2}
    elif income == '$60k - $80k':
        return {'Income_Min': 60000, 'Income_Max': 80000, 'Income_Category_Code': 3}
    elif income == '$80k - $120k':
        return {'Income_Min': 80000, 'Income_Max': 120000, 'Income_Category_Code': 4}
    elif income == '$120k+':
        return {'Income_Min': 120000, 'Income_Max': None, 'Income_Category_Code': 5}
    elif income == 'unknown':
        return {'Income_Min': None, 'Income_Max': None, 'Income_Category_Code': 6}
    else:
        return {'Income_Min': None, 'Income_Max': None, 'Income_Category_Code': None}

# 1) Apply the updated mapping function and divide the result in columns
income_mapped = df['Income_Category'].apply(map_income_category).apply(pd.Series)

# 2) Concat new columns to the DataFrame
df = pd.concat([df, income_mapped], axis=1)

# 3) Drop the original 'Income_Category' column
df.drop('Income_Category', axis=1, inplace=True)

# 4) Verify the mapping
print(df[['Income_Category_Code', 'Income_Min', 'Income_Max']].head())
```

#### Using a function, the Income_Info sub-document is created and nested within each row. This results in more structured and accessible data for improved data handling and queries.

```python
# 5) Create the 'Income_Info' sub-document,
#    handling NaN → None and int(...) conversion in the same step
def build_income_subdoc(row):
    cat_code = row["Income_Category_Code"]
    i_min = row["Income_Min"]
    i_max = row["Income_Max"]
    
    # If cat_code is NaN, use None; else int(...)
    if pd.isna(cat_code):
        cat_code = None
    else:
        cat_code = int(cat_code)
    
    # If i_min or i_max are NaN, use None; else int(...)
    if pd.isna(i_min):
        i_min = None
    else:
        i_min = int(i_min)
        
    if pd.isna(i_max):
        i_max = None
    else:
        i_max = int(i_max)
    
    return {
        "Category_Code": cat_code,
        "Min": i_min,
        "Max": i_max
    }

df["Income_Info"] = df.apply(build_income_subdoc, axis=1)

# 6) Drop the separate columns now that they are nested in "Income_Info"
df.drop(["Income_Category_Code", "Income_Min", "Income_Max"], axis=1, inplace=True)

print("\nStructure of the first 5 rows with nested Income_Info:")
print(df.head(5))
```

```python
# 1) Convert columns to int, float, or str only if they exist in each DataFrame

# --- INT CONVERSION ---
int_columns_customers = [
    "CLIENTNUM", "Customer_Age", "Dependent_count",
    "Months_on_book", "Total_Relationship_Count",
    "Months_Inactive_12_mon", "Contacts_Count_12_mon"
]
existing_int_cols_customers = [col for col in int_columns_customers if col in customers_data.columns]
customers_data[existing_int_cols_customers] = customers_data[existing_int_cols_customers].astype(int)

int_columns_transactions = [
    "CLIENTNUM"
]
existing_int_cols_transactions = [col for col in int_columns_transactions if col in transactions_data.columns]
transactions_data[existing_int_cols_transactions] = transactions_data[existing_int_cols_transactions].astype(int)

# --- FLOAT CONVERSION ---
# These fields must be double per my schema
float_columns_customers = [
    "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1",
    "Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2"
]
existing_float_cols_customers = [col for col in float_columns_customers if col in customers_data.columns]
customers_data[existing_float_cols_customers] = customers_data[existing_float_cols_customers].astype(float)

float_columns_transactions = [
    "Total_Trans_Amt", "Total_Revolving_Bal", "Credit_Limit",
    "Total_Amt_Chng_Q4_Q1", "Total_Ct_Chng_Q4_Q1", "Avg_Utilization_Ratio"
]
existing_float_cols_transactions = [col for col in float_columns_transactions if col in transactions_data.columns]
transactions_data[existing_float_cols_transactions] = transactions_data[existing_float_cols_transactions].astype(float)

# --- STRING CONVERSION ---
string_columns_customers = [
    "Attrition_Flag", "Gender", "Education_Level", "Marital_Status",
    "Card_Category"
]
existing_string_cols_customers = [col for col in string_columns_customers if col in customers_data.columns]
customers_data[existing_string_cols_customers] = customers_data[existing_string_cols_customers].astype(str)

string_columns_transactions = [
    # Don't have anything atm to convert to string 
]
existing_string_cols_transactions = [col for col in string_columns_transactions if col in transactions_data.columns]
transactions_data[existing_string_cols_transactions] = transactions_data[existing_string_cols_transactions].astype(str)

# 2) Insert data into MongoDB with exception handling
try:
    db.customers.delete_many({})
    db.transactions.delete_many({})
    
    customers_dict = customers_data.to_dict("records")
    transactions_dict = transactions_data.to_dict("records")
    
    db.customers.insert_many(customers_dict)
    db.transactions.insert_many(transactions_dict)
    
    print("Data inserted successfully into MongoDB.")
except pymongo.errors.BulkWriteError as bwe:
    print("Bulk write error occurred during data insertion:")
    for error in bwe.details['writeErrors']:
        print(f"Error at index {error['index']}: {error['errmsg']}")
except Exception as e:
    print(f"An unexpected error occurred during insertion: {e}")
```

# Customer Segmentation Pipeline

## 1) Defining and Running the Aggregation

**What**: I want to categorize customers by **Age**, **Gender**, **Education**, and **IncomeCat**, then count how many fall into each unique combination.

**Why**: This helps us see the most common demographic/financial segments and potentially tailor strategies to large or high-value groups.

**Result**: A DataFrame showing `(Age, Gender, Education, IncomeCat)` plus a `count` of how many customers match that combination, sorted by the most frequent first.


```python
# 1. Define the aggregation pipeline
pipeline = [
    {
        "$project": {
            # Give fields simpler aliases
            "Age": "$Customer_Age",
            "Gender": "$Gender",
            "Education": "$Education_Level",
            "IncomeCat": "$Income_Info.Category_Code",
            #"IncomeInfo": "$Income_Info"
        }
    },
    {
        "$group": {
            "_id": {
                "Age": "$Age",
                "Gender": "$Gender",
                "Education": "$Education",
                "IncomeCat": "$IncomeCat"
            },
            "count": {"$sum": 1},
            #"Incomeinfo":{"$push": "$IncomeInfo"}
        }
    },
    {
        # Sort results by descending count (most common segmentation first)
        "$sort": {"count": -1}
    }
]

# 2. Run the pipeline on the 'customers' collection
results = list(db.customers.aggregate(pipeline))

# 3. Convert the results to a DataFrame
df_segmentation = pd.DataFrame(results)

# flatten out the _id into separate columns:
if not df_segmentation.empty:
    df_segmentation = df_segmentation.join(pd.json_normalize(df_segmentation["_id"]))
    df_segmentation.drop(columns=["_id"], inplace=True)

# 4. Quick look at the resulting DataFrame
print(df_segmentation.head(10))
```

## 2) Plotting Customers by Income Category

**What**: We do a simple bar plot: `IncomeCat` on x-axis, total `count` on y-axis.

**Why**: Among all age/gender/education combos, we want the total number of customers in each income bracket.

**Result**: A bar chart summing up the counts for each income category code, with no error bars.


```python
#  Bar plot, analyzing how many customers in each "IncomeCat" overall

sns.barplot(data=df_segmentation, x="IncomeCat", y="count", estimator=sum, errorbar=None )
plt.title("Customer Segmentation by Income Category")
plt.xlabel("Income Category Code")
plt.ylabel("Number of Customers")
plt.show()
```

## 3) Pivot Table by IncomeCat × Gender
**What**: I took the grouped data by (Age, Gender, Education, IncomeCat, count) and pivoted it specifically by Gender (rows) and IncomeCat (columns) to get a table of sums.

**Why**: This shows how many customers fall under each gender for each income category.

**Result**: We can see if, for instance, IncomeCat=1 (Less than $40K) has more men or women, and how big those differences are.

```python
pivot_income_gender = df_segmentation.pivot_table(
    index="Gender",
    columns="IncomeCat",
    values="count",
    aggfunc="sum",
    fill_value=0
)
print("Pivot: IncomeCat vs. Gender\n", pivot_income_gender)

# 2. Simple bar chart or heatmap
pivot_income_gender.plot(kind="bar", stacked=True, figsize=(8, 5))
plt.title("Number of Customers by Gender and Income Category")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.legend(title="IncomeCat Code", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

## 4) Heatmap of Education × IncomeCat
**What**: Another pivot table, this time with Education_Level on one axis and IncomeCat on the other, summing the count.

**Why**: To quickly assess how education levels correlate with income brackets. For example, do “Graduate” customers cluster in higher income categories, or is it spread out?

**Result**: A heatmap that color-codes the cell values (the number of customers). Darker/lighter cells show bigger/smaller groups.

```python
# 1. Create a pivot table by Education (rows) and IncomeCat (columns), summing 'count'
pivot_edu_income = df_segmentation.pivot_table(
    index="Education",
    columns="IncomeCat",
    values="count",
    aggfunc="sum",
    fill_value=0
)

print("Pivot: Education vs. IncomeCat\n", pivot_edu_income)

# 2. Heatmap for visual insight
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_edu_income, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Number of Customers by Education Level and Income Category")
plt.xlabel("Income Category Code")
plt.ylabel("Education Level")
plt.show()
```

## 5) Segment By Age Ranges

**What**: I bucket Customer_Age into ranges like [18–29], [30–39], [40–49], [50–59], [60+]. Then again, we pivot or group by these AgeRange bins and IncomeCat.

**Why**: A raw “Age=43” can be too granular. Grouping into age brackets yields clearer patterns, e.g., do younger customers mostly fall into lower income brackets?

**Result**: Another pivot table or heatmap showing how many customers of each age bracket land in each income category.


```python
# I convert numeric Age -> a custom range label in Python:

df_segmentation["AgeRange"] = pd.cut(
    df_segmentation["Age"],
    bins=[18, 30, 40, 50, 60, 120],
    labels=["18-29", "30-39", "40-49", "50-59", "60+"],
    right=False
)

print("dataFrame after cut\n", df_segmentation)

# Then pivot or group by AgeRange + IncomeCat, etc.
pivot_age_income = df_segmentation.pivot_table(
    index="AgeRange",
    columns="IncomeCat",
    values="count",
    aggfunc="sum",
    fill_value=0,
    observed=False
)

sns.heatmap(pivot_age_income, annot=True, fmt="d")
plt.title("Customers by Age Range and Income Category")
plt.show()
```

# Customer Status Analysis

## 1) Defining and Running the Aggregation

**What**: I take each customer’s attrition status (`Attrition_Flag`), number of inactive months (`Months_Inactive_12_mon`), and contacts over the last 12 months (`Contacts_Count_12_mon`). Then we group by `(Attrition, MonthsInactive)` to see how many customers fall into each combination, also averaging the number of contacts.

**Why**: This helps us spot whether higher inactivity months correlate with “Attrited Customer,” and check if contact frequency differs across those segments.

**Result**: A DataFrame of `(Attrition, MonthsInactive, avgContacts, count)`, sorted by the group with the most customers first.


```python
# 1) Define the pipeline
#    I'll project only the fields I care about: Attrition_Flag, Months_Inactive_12_mon, Contacts_Count_12_mon
pipeline = [
    {
        "$project": {
            "Attrition": "$Attrition_Flag",
            "MonthsInactive": "$Months_Inactive_12_mon",
            "ContactsLast12": "$Contacts_Count_12_mon"
        }
    },
    {
        "$group": {
            "_id": {
                "Attrition": "$Attrition",
                "MonthsInactive": "$MonthsInactive"
            },
            "avgContacts": {"$avg": "$ContactsLast12"}, 
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}  # largest groups first
    }
]

# 2) Run the pipeline on the 'customers' collection
status_results = list(db.customers.aggregate(pipeline))

# 3) Convert to a DataFrame
df_status = pd.DataFrame(status_results)

# 4) Flatten the _id if not empty
if not df_status.empty:
    df_status = df_status.join(pd.json_normalize(df_status["_id"]))
    df_status.drop(columns=["_id"], inplace=True)

print(df_status.head(13))
```

**Result**: A pivot table plus a heatmap so I can visually see if one group is strongly associated with more months inactive.


```python
# 1) pivot by (Attrition vs. MonthsInactive) to see how many accounts in each combination
pivot_status = df_status.pivot_table(
    index="Attrition", 
    columns="MonthsInactive", 
    values="count", 
    aggfunc="sum",
    fill_value=0
)

print("\nPivot of Attrition_Flag vs. Months_Inactive_12_mon:")
print(pivot_status)

# 2) Plot a heatmap to visualize inactivity across attrition status
plt.figure(figsize=(10,6))
sns.heatmap(pivot_status, annot=True, fmt="d", cmap="Blues")
plt.title("Inactivity (Months) Distribution by Attrition Flag")
plt.xlabel("Months_Inactive_12_mon")
plt.ylabel("Attrition Flag")
plt.show()
```

# Financial Health Assessment

## 1) Building the Aggregation Pipeline

**What**: I want to merge (`$lookup`) each customer record with its corresponding transaction record(s) to analyze credit limit, revolving balance, and utilization.  

**Why**: Combining demographic data (Income Category) with financial fields (`Credit_Limit`, `Avg_Utilization_Ratio`) helps assess overall financial health by group. 

**Result**: A grouped set of average credit limit, revolving balance, and utilization per income category, revealing which brackets have the highest or lowest financial metrics.


```python
# 1) Define the aggregation pipeline on the 'customers' collection
pipeline = [
    {
        "$lookup": {
            "from": "transactions",
            "localField": "CLIENTNUM",     # field in 'customers'
            "foreignField": "CLIENTNUM",  # field in 'transactions'
            "as": "tx"                    # transaction data stored in this array
        }
    },
    {
        # Unwind the array so each matching transaction is on its own row
        "$unwind": "$tx"
    },
    {
        # I only need certain fields: Income_Info for income category,
        # plus credit limit, revolving balance, utilization from 'transactions'.
        "$project": {
            "CLIENTNUM": 1, # to leave it like it is   
            "IncomeCat": "$Income_Info.Category_Code",
            "CreditLimit": "$tx.Credit_Limit",
            "RevolvingBal": "$tx.Total_Revolving_Bal",
            "UtilRatio": "$tx.Avg_Utilization_Ratio",
            "Age": "$Customer_Age",        
            "Gender": 1,       # to leave it like it is           
        }
    },
    {
        # group by IncomeCat to see how credit and utilization vary across income categories.
        "$group": {
            "_id": "$IncomeCat",
            "avgCreditLimit": { "$avg": "$CreditLimit" },
            "avgRevolvingBal": { "$avg": "$RevolvingBal" },
            "avgUtilRatio":   { "$avg": "$UtilRatio" },
            "count": { "$sum": 1 }
        }
    },
    {
        # Sort by highest average credit limit, for instance
        "$sort": { "avgCreditLimit": -1 }
    }
]

# 2) Run the pipeline
results = list(db.customers.aggregate(pipeline))

# 3) Convert to a DataFrame for further analysis/plotting
df_fin_health = pd.DataFrame(results)

# Flatten out _id (IncomeCat code)
df_fin_health.rename(columns={"_id": "IncomeCat"}, inplace=True)

print("Financial Health Aggregation Results:\n", df_fin_health)
```

## 2) Plotting: Average Credit Limit by Income Category

**What**: I'll do a simple bar chart with `IncomeCat` on the x-axis, `avgCreditLimit` on the y-axis.  

**Why**: This reveals which income bracket has the highest average credit limit.  

**Result**: A quick visual distribution of credit limits across the different income categories.


```python
# 1) Simple bar chart: income category vs. average credit limit
sns.barplot(data=df_fin_health, x="IncomeCat", y="avgCreditLimit")
plt.title("Avg Credit Limit by Income Category Code")
plt.xlabel("Income Category (Code)")
plt.ylabel("Average Credit Limit")
plt.show()
```

### 1) Simple Range Query (Filtering by Age)
**Goal**: Retrieve all customers between 20 and 30 years old. Then measure how an index on `Customer_Age` might improve query efficiency.

1. Create an Index on `Customer_Age`
2. Run the Query with `.explain()`
3. Drop the Index and Run Again

**What I found:** Here we can see how much the appropriate index for this query let me to exploit MongoDB’s B-Tree–based indexing to avoid scanning the entire collection(**IXSCAN**). This balanced approach speed up the query and it's also scalable. Actually in a huge dataset perform always the same important query without an appropriate index (**COLLSCAN**) slow down the query making pymongo to check all documents everytime. 

```python
#1.1 Create the index
from pymongo import ASCENDING, DESCENDING
import time

# Create an index on Customer_Age
db.customers.create_index([("Customer_Age", ASCENDING)], name="idx_customer_age")

# List indexes for verification
print("Indexes on 'customers' collection:")
for idx in db.customers.list_indexes():
    print(idx)
```

```python
#1.2 Run the Range Query and Measure Time
# Define the filter for age between 20 and 30 (inclusive).
age_filter = {"Customer_Age": {"$gte": 20, "$lte": 30}}

start_time = time.time()
docs_with_index = list(db.customers.find(age_filter))
end_time = time.time()

print(f"Found {len(docs_with_index)} documents with index in {end_time - start_time:.6f} seconds.")

# EXPLAIN PLAN
explain_plan_with_index = db.customers.find(age_filter).explain()
print("Explain output with index:")
print(explain_plan_with_index)
```

```python
#1.3 Drop the Index and Repeat
# Drop the index
db.customers.drop_index("idx_customer_age")

start_time = time.time()
docs_no_index = list(db.customers.find(age_filter))
end_time = time.time()

print(f"Found {len(docs_no_index)} documents without index in {end_time - start_time:.6f} seconds.")

# EXPLAIN PLAN without index
explain_plan_no_index = db.customers.find(age_filter).explain()
print("Explain output without index:")
print(explain_plan_no_index)
```

### 2) Partial Matching

**Goal**: Performs a partial matching query on the Education_Level field

1) Create an index on the Education_Level field to speed up the query.
2) Define and run a query using a regular expression to perform a case‐insensitive partial match (e.g., matching "High" in "High School").
3) Use the explain plan to analyze the impact of the index on the query performance.
4) Discuss the planning and indexing strategy.

**What I Found**: In this case, with an anchored regex (^Doctorate), the query planner chooses a IXSCAN. Regex queries, not anchored ones, can't instead prevent MongoDB from effectively narrowing the index range, leading to a COLLSCAN performing as an IXSCAN.

## 3) Experiment on Indexing Performance in Aggregation Pipelines

Here I built a multi-stage aggregation pipeline that:
1) Filters for attrited customers older than 45.
2) Uses `$lookup` to join customer documents with their corresponding transaction details.
3) Unwinds the joined arrays to process each transaction individually.
4) Applies **$addFields** to extract and rename nested fields for clarity.
5) Groups the data by **Card_Category** to compute aggregate statistics (average credit limit, revolving balance, utilization, etc.).
6) Projects the final results, including the count of unique customers.
7) Sorts the output by descending average credit limit.

**Objective**

We tested the performance impact of creating a compound index on the fields used in the `$match` stage ,specifically, on `Attrition_Flag` and `Customer_Age`,versus using an index on `CLIENTNUM`. Our aggregation pipeline filters for attrited customers older than 45, joins them with their transaction details via `$lookup`, and then performs further processing.

**What I Did**

- **Without any index**: The pipeline executed in approximately **1.82 seconds**.
- **With an index on `CLIENTNUM`**: The pipeline ran significantly faster, in about **0.036 seconds**.
- **With a compound index on `Attrition_Flag` and `Customer_Age` (applied to the `$match` stage)**: The execution time was about **1.85 seconds**.
  - The `explain` plan confirmed that the compound index was used via an **IXSCAN**, indicating that MongoDB applied the index during the filtering stage.

**What I Found**

- Although the compound index on the `$match` fields is utilized (as seen in the `explain` plan), it did not improve performance in fact, the overall execution time was similar to the run without an index on those fields.
- This result suggests that in our pipeline, the benefits of a compound index on filtering are **overshadowed by other stages** (such as `$lookup`, `$unwind`, and grouping), which introduce additional overhead.
- In our case, **indexing on `CLIENTNUM`** (used for joining) yields a greater performance benefit, likely because `CLIENTNUM` is unique and the join operation is a key component of our aggregation.

**Conclusion**

The experiment shows that:

- **Compound indexes** can be effective for targeted filtering, but their impact may be limited in pipelines dominated by join and aggregation operations.
- For our specific workload, a **simple index on the join field (`CLIENTNUM`)** provides a more significant performance boost.


```python
# Define the complex aggregation pipeline
pipeline_complex = [ 
    # 1. Filter for attrited customers older than 45.
    {"$match": {
        "Attrition_Flag": "Attrited Customer",
        "Customer_Age": {"$gt": 45}
    }},
    # 2. Join customer documents with their transaction details.
    {"$lookup": {
        "from": "transactions",
        "localField": "CLIENTNUM",
        "foreignField": "CLIENTNUM",
        "as": "tx_docs"
    }},
    # 3. Unwind the array from $lookup to process each transaction separately.
    {"$unwind": "$tx_docs"},
    # 4. Add fields from nested structures for clarity and easier grouping.
    {"$addFields": {
        "IncomeCategory": "$Income_Info.Category_Code",
        "CreditLimit": "$tx_docs.Credit_Limit",
        "RevolvingBal": "$tx_docs.Total_Revolving_Bal",
        "Utilization": "$tx_docs.Avg_Utilization_Ratio"
    }},
    # 5. Group by Card_Category and calculate aggregate statistics.
    {"$group": {
        "_id": "$Card_Category",
        "avgCreditLimit": {"$avg": "$CreditLimit"},
        "avgRevolvingBal": {"$avg": "$RevolvingBal"},
        "avgUtilization": {"$avg": "$Utilization"},
        "customerIDs": {"$addToSet": "$CLIENTNUM"},
        "totalTransactions": {"$sum": 1}
    }},
    # 6. Project final fields, including the number of unique customers.
    {"$project": {
        "avgCreditLimit": 1,
        "avgRevolvingBal": 1,
        "avgUtilization": 1,
        "numCustomers": {"$size": "$customerIDs"},
        "totalTransactions": 1
    }},
    # 7. Sort the results by descending average credit limit.
    {"$sort": {"avgCreditLimit": -1}}
]

#Execution without index
start_time = time.time()
results_no_index = list(db.customers.aggregate(pipeline_complex))
time_no_index = time.time() - start_time
print(f"Execution time WITHOUT index: {time_no_index:.6f} seconds")
for doc in results_no_index:
    print(doc)
print("\n")
# Explain the aggregation pipeline WITHOUT index
explain_no_index = db.command({
    "aggregate": "customers",
    "pipeline": pipeline_complex,
    "explain": True
})
print("Explain output WITHOUT index:")
print(explain_no_index)



# Create Indexes on CLIENTNUM in both collections
db.customers.create_index([("CLIENTNUM", pymongo.ASCENDING)], name="idx_clientnum")
db.transactions.create_index([("CLIENTNUM", pymongo.ASCENDING)], name="idx_clientnum")

#Execution with index
start_time = time.time()
results_with_index = list(db.customers.aggregate(pipeline_complex))
time_with_index = time.time() - start_time
print(f"Execution time WITH index: {time_with_index:.6f} seconds")
for doc in results_with_index:
    print(doc)
print("\n")
# Explain the aggregation pipeline WITH index
explain_with_index = db.command({
    "aggregate": "customers",
    "pipeline": pipeline_complex,
    "explain": True
})
print("Explain output WITH index:")
print(explain_with_index)



# Delete index to restart the query

db.customers.drop_index("idx_clientnum")
db.transactions.drop_index("idx_clientnum")
#for idx in db.customers.list_indexes():
#    print(idx)
#for idx in db.customers.list_indexes():
#    print(idx)
```

## Advanced Aggregation with $facet: Leveraging MongoDB for Multifaceted Analysis

MongoDB's `$facet` stage allows us to run multiple aggregation pipelines concurrently within a single query. This capability enables us to compute various statistics (such as metrics for attrited customers, existing customers, and overall customer statistics) in one shot. 

In a relational database, achieving the same results would require executing several separate queries or constructing complex multi-join queries, which are harder to maintain and typically less efficient. With MongoDB, I can take advantage of its flexible schema and aggregation framework to simplify my analytics and reduce overhead.

**What I Found:**  
- The `$facet` stage lets me compute parallel sub pipelines that return diverse metrics in one operation.
- This not only simplifies the querying process but also improves performance by reducing the need to repeatedly scan the dataset.
- In contrast, similar operations in a relational database would involve multiple queries or complex subqueries, increasing both the implementation effort and processing time.


```python
pipeline_facet = [
    # Optionally, add a global match stage if needed to filter the documents to be analyzed
    {"$match": {"Customer_Age": {"$gte": 30}}},
    {"$facet": {
        "attrited_customers": [
            {"$match": {"Attrition_Flag": "Attrited Customer"}},
            {"$group": {
                "_id": "$Card_Category",
                "count": {"$sum": 1},
                "avgAge": {"$avg": "$Customer_Age"},
                "avgCreditLimit": {"$avg": "$Income_Info.Min"}  # Example: using Income_Info.Min as a placeholder
            }},
            {"$sort": {"count": -1}}
        ],
        "existing_customers": [
            {"$match": {"Attrition_Flag": "Existing Customer"}},
            {"$group": {
                "_id": "$Card_Category",
                "count": {"$sum": 1},
                "avgAge": {"$avg": "$Customer_Age"},
                "avgCreditLimit": {"$avg": "$Income_Info.Min"}  # Example: using Income_Info.Min as a placeholder
            }},
            {"$sort": {"count": -1}}
        ],
        "overall_stats": [
            {"$group": {
                "_id": None,
                "totalCustomers": {"$sum": 1},
                "avgAge": {"$avg": "$Customer_Age"}
            }}
        ]
    }}
]

# Execute the pipeline
results_facet = list(db.customers.aggregate(pipeline_facet))
pprint(results_facet)

# Explain the pipeline
explain_facet = db.command({
    "aggregate": "customers",
    "pipeline": pipeline_facet,
    "explain": True
})
pprint(explain_facet)
```
